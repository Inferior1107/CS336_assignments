```python
import pickle
import os

# 1. 设置路径
vocab_path = "../../tokenizer_outputs/vocab.pkl"
merges_path = "../../tokenizer_outputs/merges.pkl"
txt_vocab_path = "../../tokenizer_outputs/vocab_readable.txt"
txt_merges_path = "../../tokenizer_outputs/merges_readable.txt"

# 2. 读取 PKL 文件
with open(vocab_path, "rb") as f:
    loaded_vocab = pickle.load(f)

with open(merges_path, "rb") as f:
    loaded_merges = pickle.load(f)

print(f"成功读取！开始导出文本...")

# 3. 将全部 Vocab 写入 TXT
with open(txt_vocab_path, "w", encoding="utf-8") as f:
    f.write(f"=== 完整词表 (Total Size: {len(loaded_vocab)}) ===\n")
    for token_id, token_bytes in loaded_vocab.items():
        # 用 repr() 可以在文本中清晰地看到空格和特殊字符（比如 ' hello' 会显示为 b' hello'）
        # 同时尝试解码出人类可读的字符串
        try:
            readable_text = token_bytes.decode('utf-8')
            f.write(f"ID: {token_id:<5} | Text: {readable_text!r:<20} | Bytes: {token_bytes}\n")
        except UnicodeDecodeError:
            f.write(f"ID: {token_id:<5} | Text: [Decode Error]       | Bytes: {token_bytes}\n")

# 4. 将全部 Merges 写入 TXT
with open(txt_merges_path, "w", encoding="utf-8") as f:
    f.write(f"=== 完整合并列表 (Total Merges: {len(loaded_merges)}) ===\n")
    for i, (p1, p2) in enumerate(loaded_merges):
        # 尽量解码展示合并过程
        try:
            t1 = p1.decode('utf-8', errors='replace')
            t2 = p2.decode('utf-8', errors='replace')
            t_merged = (p1 + p2).decode('utf-8', errors='replace')
            f.write(f"Merge #{i+1:<5} | {t1!r} + {t2!r} -> {t_merged!r}\n")
        except Exception:
            f.write(f"Merge #{i+1:<5} | {p1} + {p2} -> {p1+p2}\n")

print(f"✅ 全部导出完毕！")
print(f"请在 VSCode 中打开查看: \n1. {txt_vocab_path} \n2. {txt_merges_path}")
```
